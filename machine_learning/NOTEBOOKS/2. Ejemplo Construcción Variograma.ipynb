# Construcción de Variograma

**SEMIVARIOGRAMA O VARIOGRAMA:** Es la herramienta geoestadística básica y expresa la correlación espacial entre los valores muestreados.

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
Image(filename="../../LIBROS Y COSAS EXTRAS/variograma.jpg")
```

```python
datos = {"Leyes":[1,2,3,4,5,4,3,2,1]}
df = pd.DataFrame(data=datos)
df
```

```python
p = []
s = []
gh = []
d = []

for i in range(len(df)-1):
    h = i+1
    if len(df)-h != 0:
        g = 1/(2*(len(df)-h))
    for j in range(len(df)):
        if j+h < len(df):
            n = (df["Leyes"][j] - df["Leyes"][j+h])**2
            p.append(n)
    
    d.append(h)
    m = sum(p)
    s.append(m)
    k = g*m
    p = []
    gh.append(k)
    
    
print(p)
print(s)
print(gh)
print(d)
```

```python
datos = {"distancia":d,"variograma":gh}
variograma = pd.DataFrame(data=datos)
variograma.plot(kind="scatter", x="distancia", y="variograma")
```
