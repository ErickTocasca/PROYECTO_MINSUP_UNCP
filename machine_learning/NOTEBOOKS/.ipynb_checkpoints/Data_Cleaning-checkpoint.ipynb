## CARGA DE DF CON FUNCIÓN OPEN

```python
data = open(mainpath + "/" + "Estimacion_Recursos/MBOK.txt", "r")
```

```python
cols = data.readline().strip().split(";")
cols
ncols = len(cols)
```

```python
ncols
```

```python
counter = 0
main_dic = {}

for col in  cols:
    main_dic[col] = []
    
    
main_dic
```

```python
for line in data:
    values = line.strip().split(";")
    for i in range(ncols):
        main_dic[cols[i]].append(values[i])
    counter += 1

print(f"El dataframe tiene {counter} filas y {ncols} columnas")
```

```python
df3 = pd.DataFrame(main_dic)
```
